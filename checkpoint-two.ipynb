## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python

#  To see the balance between employees who leave and those who stay, specifically the attrition rate in the company
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=HR_df)
plt.title('Distribution of Attrition')
plt.show()
```

```python

# To understand the relationship between job satisfaction levels and employee attrition

plt.figure(figsize=(10, 6))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=HR_df)
plt.title('Job Satisfaction vs. Attrition')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Count')
plt.show()
```

```python
# To understand the relationship between age,income and attrition

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='MonthlyIncome', hue='Attrition', data=HR_df)
plt.title('Age vs. Monthly Income')
plt.show()
```

```python

# To understand the relationship between jobrole and Attrition
plt.figure(figsize=(12, 6))
sns.countplot(y='JobRole', hue='Attrition', data=HR_df)
plt.title('Job Role vs. Attrition')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.show()
```
